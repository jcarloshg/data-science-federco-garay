Con el conjunto de datos cargado como se indica en el Ejercicio 1:


Ahora vamos a comprobar **La Estandarización de datos y Aplicamos PCA**. Para ello sigue estos pasos:

1. Crea la variable **X_centrado** y almacena el calculo de la media de cada característica (columna) en la matriz de datos X
2. Crea la variable **pca** y almacena una instancia del modelo PCA (Análisis de Componentes Principales) utilizando la clase PCA de la biblioteca sklearn.decomposition.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

iris = load_iris()
X = iris.data
y = iris.target

X_centrado = X - np.mean(X, axis=0)

pca = PCA()
X_pca = pca.fit_transform(X_centrado)
X_pca
```

# Análisis del Componentes Principales (PCA) en Machine Learning 3

**Ejercicio 3: Visualización de los resultados.**

Con los conjuntos de datos cargado y entrenados como se indica en el Ejercicio 2:

Ahora vamos a **Visualizar los resultados**. Para ello sigue estos pasos:

1. Crea una variable llamada especies y almacena la lista de especies ("setosa", "versicolor", "virginica")

2. Crea un gráfico de dispersión (scatter) empleando un loop for para iterar a través de los índices 0, 1 y 2, que corresponden a las tres especies en el conjunto de datos (setosa, versicolor y virginica, respectivamente). Dentro del loop for

   - Dentro del bucle, debes llamar a plt.scatter para crear un gráfico de dispersión.

     - X_pca[y == i, 0] selecciona todas las filas de X_pca donde la etiqueta (y) es igual a i, y toma la primera columna de esas filas (componente principal 1).
     - De forma similar, X_pca[y == i, 1] toma la segunda columna (componente principal 2).

   - label=especies[i] asigna una etiqueta a los puntos que se están graficando en esta iteración, lo cual es útil para la leyenda del gráfico.

3. Define el tamaño del gráfico en (8, 6)

Etiqueta del eje x: 'Primer componente principal'

Etiqueta del eje y: 'Segundo componente principal'

Titulo del gráfico: 'PCA del conjunto Iris'

Asegurate de mostrar la leyenda del gráfico.

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

iris = load_iris()
X = iris.data
y = iris.target

X_centrado = X - np.mean(X, axis=0)

pca = PCA()
X_pca = pca.fit_transform(X_centrado)


plt.figure(figsize=(8, 6))

for i in range(0, 3):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=iris.target_names[i])

plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid()
plt.legend()
plt.show()
```
