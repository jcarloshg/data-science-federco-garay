```python
import numpy as np

A = np.array([[1, 2], [3, 4], [5, 6]])

U, S, VT = np.linalg.svd(A)

print("U:\n", U)
print("S:\n", S)
print("VT:\n", VT)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
target = iris.target
species_name = iris.target_names
print(f"species: {species_name}")

x_centrado = X - np.mean(X, axis=0)

U, S, VT = np.linalg.svd(x_centrado)

print("U:\n", U)
print("S:\n", S)
print("VT:\n", VT)

K = 2
X_transformado = U[:, :K] * S[:K]

plt.figure(figsize=(8, 6))
for i in range(0, 3):
    plt.scatter(
        X_transformado[target == i, 0],
        X_transformado[target == i, 1],
        label=species_name[i]
    )

plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid()
plt.legend()
plt.show()
```
