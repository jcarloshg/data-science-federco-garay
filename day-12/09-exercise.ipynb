## Ejercicio 2

Con el agrupamiento jer√°rquico previamente realizado en el **Ejercicio 1**, visualiza el resultado en un dendrograma siguiendo estas indicaciones:

---

### Requisitos de la visualizaci√≥n:

1. **Tama√±o de la figura:** 10 x 7  
2. **Conteos de elementos:** Mostrar el n√∫mero de elementos originales en cada cl√∫ster en el dendrograma.  
3. **Orientaci√≥n:** El √°rbol debe crecer hacia abajo y las hojas deben estar en la parte superior.  
4. **Etiquetas de las hojas:** Asigna las etiquetas reales de las especies para cada muestra en el conjunto de datos *Iris*.  
5. **Orden de distancias:** Las distancias m√°s altas deben estar en la parte superior del dendrograma.  
6. **T√≠tulo:** `Dendrograma del Agrupamiento Jer√°rquico - M√©todo de Enlace Ward`  
7. **Eje X:** `√çndice de la Muestra`  
8. **Eje Y:** `Distancia`  

---

> üí° **Recuerda:**  
> Importa los m√≥dulos de Python necesarios para que tu c√≥digo funcione correctamente.


```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

iris = load_iris()
X = iris.data
y = iris.target

target_names = iris.target_names

linkage = linkage(X, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(
    linkage,
    orientation='top',
    labels=iris.target,
    distance_sort='descending',
    show_leaf_counts=True
)
plt.title('Dendrograma del Agrupamiento Jer√°rquico - M√©todo de Enlace Ward')
plt.xlabel('√çndice de la Muestra')
plt.ylabel('Distancia')
plt.show()
```
