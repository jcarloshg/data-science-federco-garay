```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

# Set a random seed for reproducibility
# to ensure that the results are consistent across runs
np.random.seed(42)

# Generate 100 random points in 2D space
data = np.random.rand(500, 2)
data

# train a KMeans clustering model with 5 clusters
model = KMeans(n_clusters=3, random_state=42)
model.fit(data)

```

```python
# the centers of the clusters
centroids = model.cluster_centers_
print(f"Centroids: {centroids}")

# labels for each point indicating which cluster it belongs to
labels = model.labels_
print(f"Labels: {labels}")

# Plot the data points, coloring them by their cluster label
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
import seaborn as sns

pinguinds = sns.load_dataset("penguins")
# Remove rows with missing values from the penguins dataset
pinguinds.dropna(inplace=True)
pinguinds.head()

# Select the features for clustering
new_data = pinguinds[["bill_length_mm", "bill_depth_mm"]]
new_data.head()

# Train a KMeans clustering model on the penguins dataset
pinguinds_model = KMeans(n_clusters=3, random_state=42)
pinguinds_model.fit(new_data)

# the centers of the clusters
centroids = pinguinds_model.cluster_centers_
print(f"Centroids: {centroids}")

# labels for each point indicating which cluster it belongs to
labels = pinguinds_model.labels_
print(f"Labels: {labels}")

# Plot the data points, coloring them by their cluster label
plt.figure(figsize=(10, 6))
plt.scatter(new_data["bill_length_mm"], new_data["bill_depth_mm"], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('KMeans Clustering of Penguins')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend()
plt.show()
```
