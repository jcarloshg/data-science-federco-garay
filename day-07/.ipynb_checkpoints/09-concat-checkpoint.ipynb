```python
import pandas as pd

df1 = pd.DataFrame({'Nombre': ["Juan", "Gabriela", "Elena"], 'Edad': [23, 31, 21]})
df2 = pd.DataFrame({'Nombre': ["Carmela", "Max", "Laura"], 'Edad': [34, 25, 29]})

df_concat = pd.concat([df1,df2])
df_concat
```

```python
df_concat_axis_1 = pd.concat([df1,df2], axis=1)
df_concat_axis_1
```

```python
# ignore_index -> reset the index to 1, 2, 3, 4, 5
df_concat_02 = pd.concat([df1,df2], ignore_index=True)
df_concat_02
```

```python
# ignore_index -> reset the index to 1, 2, 3, 4, 5
df_concat_02 = pd.concat([df1,df2], keys=['df1', 'df2'])
df_concat_02
```

```python
# ignore_index -> reset the index to 1, 2, 3, 4, 5
df_concat_03 = pd.concat([df1,df2], keys=['january', 'february'])
df_concat_03
```
