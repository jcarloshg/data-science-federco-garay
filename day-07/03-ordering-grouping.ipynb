```python
import pandas as pd

movie_csv_url="./movies.csv"
df = pd.read_csv(movie_csv_url)
df
```

```python
df_ordenado = df.sort_values(by="rating", ascending=False)
df_ordenado.head(10)
```

```python
df_ordenado = df.sort_values(by=['rating','recaudación(M)'], ascending=False)
df_ordenado.head(10)
```

```python
df_agrupado = df.groupby('género')['rating'].mean()
df_agrupado
```

```python
df_sum = df.groupby('año')['recaudación(M)'].sum()
df_sum.sort_values(ascending=False).head(10)
```
