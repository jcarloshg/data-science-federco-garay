```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.DataFrame()
df["Area"] = [2600, 3000, 3200, 3600, 4000]
df["Precio"] = [550000, 565000, 610000, 680000, 725000]
```

```python
#  expect a dataframe with the area and price of houses
X = df[["Area"]]
y = df["Precio"]

# create a linear regression model
modelo = linear_model.LinearRegression()
modelo.fit(X, y)
```

```python
modelo.predict([[1234]])
plt.plot(df["Area"], modelo.predict(X), color="red")
plt.scatter(df["Area"], df["Precio"])
```
