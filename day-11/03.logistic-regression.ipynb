```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv('./datos_seguro.csv')
df.head()
```

```python
plt.scatter(df.edad, df.compra, color='blue', label='Datos')
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['edad']], df['compra'], train_size=0.9)
model = LinearRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test)

new_data = pd.DataFrame({'edad': [20, 25, 35, 50, 60, 70]})
predictions = model.predict(new_data)

plt.scatter(df.edad, df.compra, color='blue', label='Datos')
plt.plot(new_data.edad, predictions, color='red', label='Predicciones')
plt.scatter(new_data.edad, predictions, color='red', label='Predicciones')
plt.xlabel('Edad')
plt.ylabel('Compra')
plt.title('Predicción de Compras por Edad')
```
