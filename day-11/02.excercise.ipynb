Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

# Crear DataFrame
data = {
    'Edad': [25, 30, 35, 40, 45],
    'Salario': [3000, 3500, 4000, 4500, 5000]
}
df = pd.DataFrame(data)

model = linear_model.LinearRegression()
X = df[["Edad"]]
y = df["Salario"]
model.fit(X, y)

plt.title('Relación entre Edad y Salario')
plt.grid(True)
plt.xlabel('Edad')
plt.ylabel('Salario')

plt.scatter(df["Edad"], df["Salario"], marker='o', color="blue")
plt.plot(df["Edad"], model.predict(X))
````
