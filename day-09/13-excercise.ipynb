Creación de Múltiples Gráficos con Matplotlib 1

Dibuja un gráfico de línea en el primer subplot (izquierdo) con los datos siguientes [1, 2, 3, 4], [2, 4, 8, 16]

Y otro gráfico scatter (derecho) en el segundo subplot. con los datos siguientes [1, 2, 3, 4], [2, 4, 8, 16]

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=2, nrows=1)

axs1 = axs[0]
axs2 = axs[1]

axs1.plot([1, 2, 3, 4], [2, 4, 8, 16])
axs2.scatter([1, 2, 3, 4], [2, 4, 8, 16])

plt.show()
````

# Creación de Múltiples Gráficos con Matplotlib 3

Personaliza los dos subgráficos del Ejercicio 2

Creación de Múltiples Gráficos con Matplotlib 2

Emplea los mismo datos:

```python
x = [1, 2, 3, 4]
y = [2, 4, 8, 16]
```

Agregando títulos diferentes a cada gráfico, cambiando el color de las líneas gráfica scatter. Asegúrate de que cada elemento de personalización se aplique correctamente al subplot correspondiente de la forma siguiente:

Gráfico 1 (Izquierda):

- Titulo: `Gráfico de línea`

- Color de línea: `verde`

Gráfico 2 (Derecha):

- Titulo: `Gráfico de scatter`

- Color de simbolos: `rojo`


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 8, 16]

fig, axs = plt.subplots(ncols=2, nrows=1)

axs1 = axs[0]
axs2 = axs[1]

axs1.plot(x, y, color='green')
axs1.set_title('Gráfico de línea')
axs2.scatter(x, y, marker='o', color='red')
axs2.set_title('Gráfico de scatter')

plt.show()
````
