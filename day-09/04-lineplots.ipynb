```python
import matplotlib.pyplot as plt
import pandas as pd
import random
%matplotlib inline

fig = plt.figure()
axs = plt.axes()


x = list(range(1,8))
y = list(range(1,8))
random.shuffle(y)

plt.plot(x,y, "ro-")
plt.grid()
plt.show()
```

```python
plt.plot(x,y, "bD:")
plt.axis([0, 10, 0, 15])
plt.grid()
plt.show()
```
