```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o')
plt.show()
```

```python
aleatorio = np.random.RandomState(0)
marcadores = ['o', 'x', '+', 'v', '^', 's', 'D']
number_items = 10
for marcador in marcadores:
    plt.plot(aleatorio.rand(number_items), aleatorio.rand(number_items), marcador)
plt.show()
```

```python
aleatorio = np.random.RandomState(0)

plt.scatter(
    aleatorio.rand(100),
    aleatorio.rand(100),
    c=aleatorio.rand(100),
    s=aleatorio.rand(100) * 1000,
    alpha=0.5,
    cmap='viridis'
    )
plt.colorbar()
plt.show()
```
