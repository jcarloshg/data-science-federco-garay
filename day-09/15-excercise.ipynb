# Estilo para tus Gráficos con Matplotlib 1

Crea un gráfico de barras utilizando Matplotlib para visualizar el número de ventas de diferentes tipos de frutas. Utiliza el siguiente conjunto de datos:

```python
df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],

```

1. Asegurate de que en el eje X se muestren las frutas, y en el eje Y se muestren las cantidades.

2. Añade al gráfico, el siguiente título: 'Distribución de frutas en el almacén'


````python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})

plt.bar(df['frutas'], df['cantidades'])
plt.title('Distribución de frutas en el almacén')

plt.show()
````

# Estilo para tus Gráficos con Matplotlib 3

Basado en los datos de frutas y ventas que hemos venido trabajando en los dos ejercicios anteriores:

```python
df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})
```

1. Crea un gráfico de pastel que muestre la distribución de frutas en el almacén

2. El gráfico de pastel debe incluir etiquetas que identifiquen cada tipo de fruta.

3. Además, usa el parámetro correcto para mostrar el porcentaje de cada fruta en el gráfico en el siguiente formato:

   - Tipo float.
   - Un solo decimal después de la coma
   - Coloca el simbolo de % al final

4. Separa todas las poriciones, es decir explotalas todas para que se vean separadas.

5. Añade al gráfico el siguiente título: 'Distribución de frutas en el almacén'

6. Finalmente agrega el estilo "dark_background" a tu gráfico.

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.


````python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})

explode = (0.1, 0.1, 0.1, 0.1)  # Explode all slices for emphasis

plt.style.use('dark_background')

plt.pie(
    df['cantidades'],
    labels=df['frutas'],
    autopct='%1.1f%%',
    explode=explode
)

plt.title('Distribución de frutas en el almacén')

plt.show()
````
