# Gráficos Scatter en Matplotlib 1

Crea un gráfico scatter simple usando Matplotlib para visualizar la relación entre dos listas de datos:

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

Utiliza el método scatter() para generar el gráfico.

NO es necesario el uso de inline en esta plataforma %matplotlib inline


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y)
plt.show()
````

# Gráficos Scatter en Matplotlib 3

Genera un gráfico scatter que muestre dos series de datos.

Para la primera serie utiliza:

Color: azul

Marcador: 'o'

Etiqueta: Serie 1

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
```

Para la segunda serie emplea:

Color: verde

Marcador: '^'

Etiqueta: Serie 2

```python
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
```


````python
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]

x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.scatter(x1, y1, color='blue', marker='o', label='Serie 1')
plt.scatter(x2, y2, color='green', marker='^', label='Serie 2')
plt.legend()
plt.show()
````
