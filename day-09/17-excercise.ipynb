# Crea Cualquier tipo de Gráficos desde Matplotlib 1

Investiga en la documentación de Matplotlib y crea un gráfico de tallo (stem):

Que represente la relación entre dos conjuntos de datos `x` e `y`.

Los valores para x deben ser los números enteros del 1 al 10, y los valores para y deben ser el cuadrado de cada número en x.

Asegúrate de incluir títulos adecuados para el gráfico, así como para los ejes x e y.

- Título del gráfico: "Relación entre Números y sus Cuadrados"

- Título del eje x: "Número (x)"

- Título del eje y: "Cuadrado del número (x^2)"

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.


````python
import matplotlib.pyplot as plt

x = list(range(1, 11))
y = [i**2 for i in x]

fig, ax = plt.subplots()

ax.stem(x, y)
ax.set_title('Relación entre Números y sus Cuadrados')
ax.set_xlabel('Número (x)')
ax.set_ylabel('Cuadrado del número (x^2)')

plt.show()
````

# Crea Cualquier tipo de Gráficos desde Matplotlib 2

Investiga en la documentación de Matplotlib y crea un gráfico apilado (stackplot) con los siguientes datos:

- y1 con valores [6, 1, 4, 4, 8]
- y2 con valores [4, 6, 3, 5, 8]
- y3 con valores [7, 9, 9, 2, 7]

El eje x debe representar el tiempo, con valores del 1 al 5 y debe ser almacenado en una variable llamada x

Documentación de referencia:

stackplot(x, y)

Stackplots and streamgraphs

Asigna los siguientes colores a cada serie: y1 rojo, y2 azul, y3 verde

Agrega una leyenda que indique qué serie corresponde a y1: Serie y1, y2: Serie y2 y y3: Serie y3.

Incluye un título al gráfico: "Distribución acumulativa de valores por serie de tiempo"

Agrega etiquetas para los ejes x: "Tiempo" y y: "Valores Acumulados".

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.


````python
import matplotlib.pyplot as plt

y1 = [6, 1, 4, 4, 8]
y2 = [4, 6, 3, 5, 8]
y3 = [7, 9, 9, 2, 7]
x = range(1, 6)  # Tiempos del 1 al 5

fig, ax = plt.subplots()

ax.stackplot(
    x, y1, y2, y3,
    labels=['Serie y1', 'Serie y2', 'Serie y3'],
    colors=['red', 'blue', 'green']
)

ax.legend(loc='upper left')

ax.set_title('Distribución acumulativa de valores por serie de tiempo')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Valores Acumulados')

plt.show()
````

# Crea Cualquier tipo de Gráficos desde Matplotlib 3

Usa Matplotlib y NumPy para generar un violin plot:

violinplot(D)

Que represente la distribución de tres diferentes conjuntos de datos.

Los parámetros para cada muestra son los siguientes:

- Muestra 1: media = 0, desviación estándar = 0.5

- Muestra 2: media = 2, desviación estándar = 1

- Muestra 3: media = -2, desviación estándar = 1.5

Cada conjunto debe ser generado con la función numpy.random.normal y debe contener 100 puntos.

```python
# Generar los datos para cada muestra usando numpy.random.normal
np.random.seed(0) # Establecer la misma semilla para la generación de datos
data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)
```

1. Asegúrate de establecer etiquetas en el eje x para identificar cada una de las muestras 'Muestra 1', 'Muestra 2', 'Muestra 3'
2. Añade un título al gráfico: "Comparación de distribuciones en Violin Plot".

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.

NO es necesario el uso de inline en esta plataforma %matplotlib inline


````python
import matplotlib.pyplot as plt
import numpy as np

# Generar los datos para cada muestra usando numpy.random.normal
np.random.seed(0)  # Establecer la misma semilla para la generación de datos
data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)
data = [
    data1,
    data2,
    data3
]

fig, ax = plt.subplots()
vp = ax.violinplot(data)

# Establecer las etiquetas del eje x
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Muestra 1', 'Muestra 2', 'Muestra 3'])

# Añadir título al gráfico
ax.set_title('Comparación de distribuciones en Violin Plot')

plt.show()
````
