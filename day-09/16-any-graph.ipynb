```python
import pandas as pd
import matplotlib.pyplot as plt

lights_csv_url = './Iluminacion.csv'
df = pd.read_csv(lights_csv_url)

fig, ax = plt.subplots()

hb = ax.hexbin(
    df['Latitud'],
    df['Longitud'],
    C=df['Luminosidad'],
    gridsize=50,
    cmap='Greens',
)

ax.set_facecolor('black')
plt.colorbar(hb, label='Luminosidad')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.title('Mapa de Calor de Iluminaci√≥n en la Ciudad')
```
