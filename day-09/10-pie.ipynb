```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "etiquetas":['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    "cantidad":[30, 60, 80, 20]
})
df

plt.pie(df['cantidad'])
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'])
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%', colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'], shadow=True)
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%',
        colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'], shadow=True,
        explode=(0.1, 0, 0, 0.5))  # Explode the first slice
plt.title('Distribucin de Frutas')
plt.show()
```
