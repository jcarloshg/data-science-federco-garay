```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.displot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
)
```

```python
sns.histplot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
)
```

```python
sns.displot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
    col="species",
)
```

```python
grap = sns.relplot(
    data=penguins,
    x="flipper_length_mm",
    y="bill_length_mm",
    hue="species",
    col="sex",
)
grap.set_axis_labels(
    "Flipper Length (mm)",
    "Bill Length (mm)"
)
```
