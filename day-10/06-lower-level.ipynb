```python
import seaborn as sns
pinguinos = sns.load_dataset("penguins")
pinguinos
```

```python
chart = sns.PairGrid(
    data=pinguinos,
    hue="species",
)
```

```python
chart = sns.PairGrid(
    data=pinguinos,
    hue="species",
    corner=True,
)
```

```python
chart = sns.PairGrid(
    data=pinguinos,
    hue="species",
    corner=True,

)
chart.map_lower(sns.kdeplot, hue=None, levels=5)
```

```python
chart = sns.PairGrid(
    data=pinguinos,
    hue="species",
    corner=True,

)
chart.map_lower(sns.kdeplot, hue=None, levels=5)
chart.map_lower(sns.scatterplot, marker="+")
```

```python
chart = sns.PairGrid(
    data=pinguinos,
    hue="species",
    corner=True,

)
chart.map_lower(sns.kdeplot, hue=None, levels=5)
chart.map_lower(sns.scatterplot, marker="+")
chart.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
```

```python
chart = sns.PairGrid(
    data=pinguinos,
    hue="species",
    corner=True,

)
chart.map_lower(sns.kdeplot, hue=None, levels=5)
chart.map_lower(sns.scatterplot, marker="+")
chart.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
chart.add_legend(frameon=True)
```

```python
chart = sns.PairGrid(
    data=pinguinos,
    hue="species",
    corner=True,

)
chart.map_lower(sns.kdeplot, hue=None, levels=5)
chart.map_lower(sns.scatterplot, marker="+")
chart.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
chart.add_legend(frameon=True)
chart.legend.set_bbox_to_anchor((0.61, 0.6))
```
