```python
import seaborn as sns
import matplotlib.pyplot as plt

pinguins = sns.load_dataset("penguins")
pinguins
```

```python
sns.relplot(
    data=pinguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
)
```

```python
sns.set_theme(style="dark")
sns.relplot(
    data=pinguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    marker='x',
    s=100
)
```

```python
sns.set_theme(style="dark")
g = sns.relplot(
    data=pinguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    marker='x',
    s=100
)
g.set_axis_labels("Bill Length (mm)", "Bill Depth (mm)")
g.legend.set_title("Body Mass (g)")
```

```python
sns.set_theme(style="dark")
g = sns.relplot(
    data=pinguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    marker='x',
    s=100
)
g.set_axis_labels("Bill Length (mm)", "Bill Depth (mm)")
g.legend.set_title("Body Mass (g)")
g.figure.set_size_inches(10, 5)
```
