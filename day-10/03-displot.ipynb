```python
import seaborn as sns
pinguins = sns.load_dataset("penguins")
pinguins

```

```python
sns.displot(
    data=pinguins,
    x="body_mass_g"
)
```

```python
sns.displot(
    data=pinguins,
    x="body_mass_g",
    kde=True,
)
```

```python
sns.displot(
    data=pinguins,
    x="body_mass_g",
    kde=True,
    col="island",
)
```

```python
sns.displot(
    data=pinguins,
    x="flipper_length_mm",
    kind="kde",
    rug=True,
)
```

```python
sns.displot(
    data=pinguins,
    x="flipper_length_mm",
    kind="ecdf"
)
```
